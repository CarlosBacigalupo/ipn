```python
#! /usr/bin/env python

import subprocess

def make_query():
    """
        Returns a dictionary of parameters to fill a VizieR query using the 
        vizquery program (see. http://cdsarc.u-strasbg.fr/doc/vizquery.htx for
        more info)

        Parameters
        --------

        None

        Returns
        --------

        query : dictionary
            query is a dictionary of all the parameters for the VizieR query

    """

    query = {}

    query['object'] = 'HD 44179'            # object can use ra and dec just as
                                            # you would in VizieR 
                                            # i.e. '06 19 58.21 -10 38 14.69'
                                            # would suffice

    query['radius'] = '5.0'                 # radius of the target                               

    #query['max'] = '1.0'                    # maximum number of lines to be return

    query['source'] = '2mass'               # catalogue to query. can also use
                                            # the VizieR catalogue names 
                                            # i.e. 'II/246/out' for the 2MASS All-Sky Catalogue
                                            # of Point Sources (Cutri+ 2003)

    query['list'] = 'list.txt'
    
    query['output'] = 'Jmag Hmag Kmag'      # the columns that you actually want
                                            # from the catalogue errors are
                                            # usually just 'e_' appended to the
                                            # begining of any measurement i.e.
                                            # e_Jmag for the error in the J-band
                                            # magnitude

    return query

def not_unit(row):
    """
        check if line contains a unit
    """

    units = ['Jy', 'mag', 'mJy', 'cW/m2/nm']

    for unit in units:
        if unit in row:
            return False

    return True

def get_data(q_params):
    """
        queries VizieR using the cdsclient and returns the twomass data
    """

# this string is the query you would make from the terminal where the inputs are
# in the {}, if you want to query a list of objects -source becomes -list={list}
# and so 'list' will be needed to be added to the parameter dictionary
    query_string = "vizquery -source='{source}' -c='{object}' -c.rs='{radius}' -out='{output}' -sort='_r' -out.max='{max}'"

    #query_string = "vizquery -list='{list}' -c='{object}' -c.rs='{radius}' -out='{output}' -sort='_r' -out.max='{max}'"

# this inserts the parameters stored in the dictionary q_params into the
# query_string
    query = query_string.format(**q_params)

# here we called the vizquery program with the inputs specified in the q_params
# dictionary and then pipe the output from stdout
    (output, err) = subprocess.Popen(query, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, shell=True).communicate()

# the output is a single string with new line characters, '\n' separating each
# line this command separates each line into a list of lines
    rows = output.split('\n')

# after we have the data in a list of lines we now read through row by row
# taking only the rows of data
    for row in rows:
        # checks if the row is not empty '(row)', ignores commented lines '#'
        # and crap lines '----', lines with units in them you may need to add
        # more units to the list in the 'not_unit' function, and finally ignores
        # the line with all the column names 'q_params['output'].split()[0]'
        if (row) and ('#' not in row) and ('----' not in row) and (not_unit(row)) and (q_params['output'].split()[0] not in row):
            # returns a list of only the data in the form of the defined
            # 'output' in the query dictionary
            #********************************************
            # when using a list input this part will need to change to get all
            # the lines of data and not just the first one
            # *******************************************
            return row.split()

    return None

def main():
    """
        main function for testing
    """

# gets the query dictionary (need to change parameters in this dictionary to
# make different queries)
    q_params = make_query()

# gets the first line of actual data from the output
    data = get_data(q_params)

# prints the data to the screen
    print q_params['object']
    print q_params['output']
    print data

if __name__ == '__main__':
    main()
```
