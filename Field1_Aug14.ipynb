```python
giant_mask = ((a[:,4]<4) & (a[:,3]-a[:,4]>0.5))
giants = a[giant_mask]
print np.sum(giant_mask)
```

```python
#searvch for the best fitted field that includes 1 giant and most >8 mag stars
maxN = 0 
# for j in a[:,[16,17]][np.argsort(a[:,3])][:5]:
for j in giants[:,[16,17]]:
    obs_field_centre=j
    for i in range(0,360,10):
        for r in np.arange(0,1,0.1):
            thisCenter = (r*np.sin(np.pi/180*i)+obs_field_centre[0], r*np.cos(np.pi/180*i)+obs_field_centre[1])
            plt.scatter(thisCenter[0],thisCenter[1], marker='+', c='g')
            target_stars_mask = ((a[:,16]-thisCenter[0])**2+(a[:,17]-thisCenter[1])**2)<=1
            mag_mask = a[:,4]>8
            if np.sum(target_stars_mask & mag_mask)>=maxN :
#                 print 'New max', np.sum(target_stars_mask & mag_mask), thisCenter
                maxN = np.sum(target_stars_mask & mag_mask)

plt.scatter(thisCenter[0],thisCenter[1], label='Tested target centres', marker='+', c='g')                
plt.scatter(a[:,16],a[:,17], s=10**(a[:,3]/7), c='b', alpha=0.6)
plt.scatter(giants[:,16],giants[:,17], s=10**(giants[:,3]/3), c='r',  label = 'Clump giants' )                
plt.xlabel('RA [deg]')
plt.ylabel('Dec [deg]')
plt.gca().invert_xaxis()
plt.title('Hyades Field - Target centres')
plt.legend()
plt.show()
```

```python
#_r|_RAJ2000|_DEJ2000|3UC|RAJ2000|DEJ2000|ePos|f.mag|ot|db|pmRA|pmDE|Jmag|Kmag|catflg
hya_ucac3 = np.genfromtxt('hya_30selected_ucac3.tsv', delimiter='|', comments='#')
hya_ucac3=np.delete(hya_ucac3,np.arange(len(hya_ucac3))[np.isnan(hya_ucac3[:,0])],0)
```

```python
#export into fld form
for i in xrange(hya_ucac3.shape[0]):
    RA1, RA2, RA3 = toolbox.dec2sex(hya_ucac3[i,1]/15)
    Dec1, Dec2, Dec3 = toolbox.dec2sex(hya_ucac3[i,2])
    thisMag = hya_ucac3[i,13]+2
    print ('HYA'+"%02d"%(i+1)).ljust(11) ,int(RA1), int(RA2), '{0:4.1f}'.format(RA3),' ', int(Dec1),  '{0:2.0f}'.format(Dec2), '{0:4.1f}'.format(Dec3), ' ','P 8 ',  '{0:5.2f}'.format(thisMag), '0','2014'
```
