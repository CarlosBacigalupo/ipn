```python
from pymodelfit.builtins import VoigtModel
from pymodelfit.builtins import GaussianModel
from pymodelfit import get_model_instance
import pylab as plt
import numpy as np
```

```python
a = get_model_instance('Voigt')
```

```python
A = 5.93304790e+04
sigma = 1.86903259
gamma = -2.75172270e-01
mu = 6.60696457e+02
A2 = 5.93304790e+04
sigma2 = 1.86903259
gamma2 = 2.75172270e-01
mu2 = 6.60696457e+02
```

```python
curve = a.f(np.arange(4000),A, sigma, gamma, mu)
curve2 = a.f(np.arange(4000),A2, sigma2, gamma2, mu2)
```

```python
plt.plot(curve)
plt.plot(curve2)
plt.show()
```
